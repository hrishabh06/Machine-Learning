```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing dataset

dataset = pd.read_csv('Position_Salaries.csv')
print(dataset)
X = dataset.iloc[:,1:-1].values
Y = dataset.iloc[:,-1].values
print("X = ",X)
print("Y = ",Y)
```

```python
# Training the Linear Regression model on whole dataset

from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,Y)
```

```python
# Training the Polynomial Regression model on whole dataset

from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly,Y)
```

```python
#Visualising the Linear & Polynomial Regression results

plt.scatter(X,Y,color = 'red')
plt.plot(X, lin_reg.predict(X), color = 'blue')             #----- Linear Regression
plt.plot(X, lin_reg_2.predict(X_poly), color = 'green')     #----- Polynomial Regression
plt.title('Truth vs Bluff (Linear & Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
#predicting the salary for new input
linear_reg_slary = lin_reg.predict([[6.5]])
poly_reg_salary = lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))

print("Linear Regression Salary Result : ", linear_reg_slary[0])
print("Polynomial Regression Salary Result : ", poly_reg_salary[0])
```
